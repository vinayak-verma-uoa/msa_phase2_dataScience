```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

%matplotlib inline
```

# Exploratory Data Analysis.

```python
dataset = pd.read_csv("weather-data.csv")
dataset.head() # Show the first five results.
```

```python
'''
Display average, standard deviation and percentile range (25th and 75th).
'''
dataset.describe()
```

```python
# Correlation Plot
sns.heatmap(dataset.corr(), annot=True, square=True)
```

```python

def plotGraphs(column_title):
    # Plot style and sizing
    plt.style.use("bmh")
    f = plt.figure()
    f.set_figwidth(10)
    f.set_figheight(10)

    # Setting up plot for temprature data over time
    plt.plot(dataset[column_title])
    plt.title('{} vs time graph'.format(column_title))
    plt.xlabel('time')
    plt.ylabel(column_title)
    # Show plot 
    plt.show()

# data columns to plot
data_columns = ['tmpc', 'dwpc', 'sknt']
for column in data_columns:
    plotGraphs(column)
```

## Comments on the dataset
* Data is quite big and needs a big plot to accuratly plot all data points. Thinning down the data by removing every third data point might help solve this.
* The correlation plot indicates a strong relationship between environment temprature and environment dew point temprature.
* The numerical section plots display proportionality between environment temprature and environment dew point temprature. Evident by the similar sinusoidal wave pattern in both plots.
* Wind direction data is incomplete.
